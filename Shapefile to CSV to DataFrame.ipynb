```python
import matplotlib.pyplot as plt
import numpy as np

# Numbers of Political Votes in 2018


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ( 'Balsam Twp', 'Beaver Twp', 'Ball Bluff Twp', 'Aitken Twp')
y_pos = np.arange(len(people))
performance = ('20', '35', '154', '474') 

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('MNAG Total Votes')
ax.set_title('2018 Voting in Minnesota')

plt.show()
```
